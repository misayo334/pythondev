```python
class Baseball_Team: 
    def __init__(self, name, win = 0, lose = 0, draw = 0): 
        self.name = name 
        self.win = win 
        self.lose = lose 
        self.draw = draw 
        
    def calc_win_rate(self): 
        self.winlose = self.win + self.lose 
        self.rate = self.win / self.winlose 
    
    def show_team_result(self): 
        self.calc_win_rate() 
        #print(self.name, self.win, self.lose, self.draw, self.rate)
        print(f"{self.name:10s}{self.win:>5d}{self.lose:>5d}{self.draw:>5d}{self.rate:8.3f}") #整列したもの
        
```

```python
archers = Baseball_Team('Archers',77,64,2)
meteors = Baseball_Team('Meteors',71,69,3)
cheetahs = Baseball_Team('Cheetahs',69,68,6)
sharks = Baseball_Team('Sharks',70,70,3)
pythons = Baseball_Team('Pythons',68,73,2)
owls = Baseball_Team('Owls',59,82,2)

#print(f"team      win  lose draw rate")  #見出し
print(f"{'team':10s}{'win':>5s}{'lose':>5s}{'draw':>5s}{'rate':>8s}") #整列した見出し

archers.show_team_result()
meteors.show_team_result()
cheetahs.show_team_result()
sharks.show_team_result()
pythons.show_team_result()
owls.show_team_result()

#ループでも実行してみる
teams = [archers, meteors, cheetahs, sharks, pythons, owls]
print("")  #改行
print(f"{'team':10s}{'win':>5s}{'lose':>5s}{'draw':>5s}{'rate':>8s}") #整列した見出し
for t in teams:
    t.show_team_result()


```
