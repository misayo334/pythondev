```python
price_wo_tax = int(input("税抜き価格を入力してください： "))   
print(f"税抜き価格を入力してください：{price_wo_tax:,}円")

#print(f"税抜き価格：{price_wo_tax:,}円")
price_w_tax = round(price_wo_tax * 1.1)
#print(f"税込み価格：{int(price_w_tax):,}円")

if price_w_tax >= 2000:

  print("送料は無料です。")
  price_w_shipping = price_w_tax

else:
   print("送料として350円かかります。")
   price_w_shipping = price_w_tax + 350

print(f"送料込みの価格は{price_w_shipping:,}円です。")
```
