```python
try: 
    a = input("割られる数を入力してください：")
    b = input("割る数を入力してください：")

    print("割られる数を入力してください：",a)
    print("割る数を入力してください：",b)

    a = int(a)
    b = int(b)
    c = a / b


except ValueError:
    print("エラー：数値を入力してください")

except ZeroDivisionError:
    print("エラー：0で割り算しないでください")

else:
    print(f"{a} ÷ {b} = {c}")

finally:
    print("処理を終了します")
```
