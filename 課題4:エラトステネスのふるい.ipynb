```python
for i in range(1,100):

    i += 1
    #print("これからこの数字が素数かチェック:",i)
    prime_check = True
 
    for j in range(1,i-1):
        
        j += 1
        #print("この数字で割り切れるかチェック：",j)

        if i % j == 0:    #割り切れたら素数ではない
        #    print("素数ではない")
            prime_check = False
            break
        
    if prime_check == True:
    #    print("この数字は素数:",i)
        print(i)
    else:
    #    print("この数字は素数ではない:",i)
        continue

        
    




```
