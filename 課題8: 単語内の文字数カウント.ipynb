```python
#英単語を格納する空のリストを作成
check_these_alphabets = []  #入力されたアルファベットを格納するリスト
list_of_words = []  #入力された英単語を格納するリスト

#キーがアルファベット各小文字で値が0の辞書を作成
alphabet_count = dict.fromkeys(list("abcdefghijklmnopqrstuvwxyz"),0)
# print(alphabet_count)

#英単語を入力してもらう
while True:
    input_word = input("英単語を入力してください（未入力＋エンターキーで終了）：")
    if input_word == "":
        print(f"英単語を入力してください：")  #no entry
        break
    else:
        check_these_alphabets += input_word
        list_of_words.append(input_word)

        print(f"英単語を入力してください：{input_word}")

#入力された英単語を昇順に並び替え
# print(f"入力された英単語（入力順）: {list_of_words}")  
list_of_words.sort()
print(f"入力した英単語: {list_of_words}") 

#入力されたアルファベットのリストの中の各文字数をカウントして辞書を更新
# print(f"入力されたアルファベット：{check_these_alphabets}")
try:  
    for s in check_these_alphabets:
        alphabet_count[s] += 1
    # print(alphabet_count)
except:     #英単語以外のものが入力された場合もエラーを出さないようにしておく
    print("入力された値の中に英単語ではないものがあったため、正常に処理できませんでした")

#カウントしたアルファベット数を表示
for t in alphabet_count:
    if alphabet_count[t] == 0:
        #print(f"{t}はゼロ")
        continue
    else:
        print(f"{t}が{alphabet_count[t]}個ありました")
```
