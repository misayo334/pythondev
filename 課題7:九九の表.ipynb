```python
#@title 九九のリストを作成する

kuku_add = []
kuku_add_alt = []
# print(kuku_add)

#二重ループで作成
for x in range(1,10):
    kuku_inner = []

    for y in range(1,10):
        kuku_inner.append(x * y)
        # print(kuku_inner)
    kuku_add.append(kuku_inner)
    
    #print(kuku_add)

print(kuku_add) #確認のため中身をチェック

#内包表記でも作成してみる
kuku_add_alt = [[(x * y) for y in range(1,10)]for x in range(1,10)]

print(kuku_add_alt) #確認のため中身をチェック
```
