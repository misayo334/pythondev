```python
score = int(0)
count = int(0)
score_sum = int(0)
average = float(0)


# while score != -1:

while True:
    if score == -1:
        break

    score = int(input("点数を入力してください： "))   
    print(f"点数を入力してください：{score:,}")

    count += 1
    score_sum += score

count -= 1    #最後の-1の入力分を戻す
score_sum += 1  #最後の-1の入力を戻す

#print(f"{count}人のテストの合計点は{score_sum}点です。")
average = score_sum / count
print(f"{count}人のテストの平均点は{average:.1f}点です。")

    



```
